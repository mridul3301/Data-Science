```python
import numpy as np
import matplotlib.pyplot as plt
```

Let's look at a simple example:

```python
rgb = np.array([[[255, 0, 0], [255, 0, 0], [255, 0, 0]],
               [[0, 255, 0], [0, 255, 0], [0, 255, 0]],
               [[0, 0, 255], [0, 0, 255], [0, 0, 255]]])

plt.imshow(rgb)
plt.show()
```

Each 2D array represents the pixel of row colors. Each element of every 2D array
consists of same pixels. SO, there is a single color throught the row.
And the collection of all the pixels and 2D rows makes a 3D array.

```python
rgb = np.array([[[255, 0, 0], [255, 255, 0], [255, 255, 255]],
               [[255, 0, 255], [0, 255, 0], [0, 255, 255]],
               [[0, 0, 0], [0, 255, 255], [0, 0, 255]]])

plt.imshow(rgb)
plt.show()
```

Let's load an RGB array and plot it to see  what it contains.

Numpy arrays can be saved in many formats such as *__.csv, .txt, .pkl, .npy__*

*.npy* is considered best.

```python
#Loading a ".npy" file.
with open("rgb_array.npy", "rb") as f:
       rgb_array = np.load(f)
plt.imshow(rgb_array)
plt.show()
```

#### Examining RGB data

```python
red_array = rgb_array[:, :, 0]
blue_array = rgb_array[:, :, 1]
green_array = rgb_array[:, :, 2]
```

```python
red_array[1].max(), green_array[1].max(), blue_array[1].max()
```

#### Updating RGB data

```python
#Replacing 145 by 30
new_array = rgb_array *0.5
new_dark_array = new_array.astype(int)
plt.imshow(new_dark_array)
plt.show()
```

Not much changed but still little bit of change can be seen.

#### Saving arrays as .npy files

```python
with open("new_array.npy", "wb") as f:
    np.save(f, new_dark_array)
```

We can now load, update and save arrays as *.npy* files.
