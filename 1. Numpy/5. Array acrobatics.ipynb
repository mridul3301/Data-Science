```python
import numpy as np
import matplotlib.pyplot as plt
```

In ML, Data Augmentation is the process of adding additional data by performing small manipulation on the data that is already available.

For Example: Let's say that we are training a model to predict whether a bottle is recylable or not using 1000 images We can use augmentation to flip or rotate image across various axis and train our model including augmented images also in order to make sure the model understands that the orientation of bottle doesnot matter.


#### Flipping array elements using *np.flip*

```python
#Loading a ".npy" file.
with open("rgb_array.npy", "rb") as f:
       normal_array = np.load(f)
plt.imshow(normal_array)
plt.show()
```

```python
#Flippin the image
flipped_array = np.flip(normal_array)
plt.imshow(flipped_array)
plt.show()
```

#### Flipping array elements along an axis

```python
#Order of the rows of pixel is reversed when axis argument is supplied with 0
flipped_rows_array = np.flip(normal_array, axis = 0)
plt.imshow(flipped_rows_array)
plt.show()
```

```python
#only rgb value can be flipped by supplying axis argument equal to 2.
flipped_rgb_array = np.flip(normal_array, axis = 2)
plt.imshow(flipped_rgb_array)
plt.show()
```

```python
#We can also flip along some axis and not along other by supplying axis as a tuple.
#Flipping row and column keeping rgb constant
new_flipped_array = np.flip(normal_array, axis = (0, 1))
plt.imshow(new_flipped_array)
plt.show()
```

```python
#Difference between flip and transpose

array = np.array([[1.1, 1.2, 1.3],
                  [2.1, 2.2, 2.3], 
                  [3.1, 3.2, 3.3],
                  [4.1, 4.2, 4.3]])
```

```python
np.flip(array)
```

```python
np.transpose(array)
```

```python
#Transpose changing axes order from (0, 1, 2) to (1, 0, 2)
transpose_array = np.transpose(normal_array, axes = (1, 0 ,2))
plt.imshow(transpose_array)
plt.show()
```
