### Regression Plots in Seaborn

Distribution anaysis was univariate because we were looking a single variable  but Regression analysis is bivariate because we are looking for relationships between two variables.

Basis for building regression plots is : *__regplot__*

It is similar to *__displot__* but we explicitly define the *dataframe*, *x* and *y*

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in the DataFrame
df = pd.read_csv("https://assets.datacamp.com/production/repositories/2210/datasets/1a8176dc594fc0a13a9f1a7b207d30ed312f2e4a/insurance_premiums.csv")
```

```python
df.head()
```

```python
#starting with a simple regplot
# Create a regression plot of premiums vs. insurance_losses
sns.regplot(data = df, x = "insurance_losses", y = "premiums")


# Display the plot
plt.show()
```

The process of plotting multiple graphs while changing a single variable is often called faceting. We can also accomplished faceting by using the *__lmplot()__* function. The base function is very similar to *__regplot()__* but it provides much more power by allowing us to add additional information using columns, colors, or rows.

Lower level regplot() and Higher level lmplot() are related. They are called the same way and produce similar output. However, the lmplot() is much more powerful.  The use of plotting multiple graphs while changing a single variable is often called faceting. 

```python
#lmplot example
# Create an lmplot of premiums vs. insurance_losses
sns.lmplot(data = df, x = "insurance_losses", y = 'premiums')


# Display the second plot
plt.show()
```

There is not much of a difference between plots generated by *regplot* and *lmplot* other than aspect ratio.

The use of *__hue__* is a powerful concept that is present throughout many of Seaborn's functions that allows faceting.

```python
# Create a regression plot using hue
sns.lmplot(data=df,
           x="insurance_losses",
           y="premiums",
           hue="Region")

# Show the results
plt.show()
```

Other concept similar to hue are *__col__* and *__row__* which produces different plot for different type based on the column passed.

```python
# Create a regression plot with multiple rows
sns.lmplot(data=df,
           x="insurance_losses",
           y="premiums",
           row="Region")

# Show the plot
plt.show()
```

```python
# Create a regression plot with multiple rows
sns.lmplot(data=df,
           x="insurance_losses",
           y="premiums",
           col="Region")

# Show the plot
plt.show()
```
