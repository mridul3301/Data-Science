Categorical data is the data that takes on a limited an fixed number of values.

```python
# import all modules
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read in the DataFrame
df = pd.read_csv("https://assets.datacamp.com/production/repositories/2210/datasets/205443d734f177d36dad2f0bdf821a57b2c4cc13/schoolimprovement2010grants.csv")
```

Seaborn breaks categorical data plots into 3 groups: 
1. Plots showing all of the individual observations : *__stripplot() and swarmplot().__*
2. Plots showing abstract representation of categorical data : *__violinplot(), boxplot() and boxenplot().__*
3. Plots showing statistical estimates of categorical values: *__barplot(), countplot() and pointplot().__*

### Plot - stripplot

```python
# Create the stripplot
sns.stripplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         jitter=True)

plt.show()
```

### Plot - swarmplot

```python
# Create and display a swarmplot with hue set to the Region
sns.swarmplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         hue='Region',
             s = 1.5)
#argument "s" is to denote size of markers in the plot
plt.show()
```

### Plot - boxplot

```python
# Create a boxplot
sns.boxplot(data=df,
         x='Award_Amount',
         y='Model Selected')

plt.show()
plt.clf()
```

### Plot - violinplot

```python
# Create a violinplot with the husl palette
sns.violinplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         palette='husl'
        )

plt.show()
plt.clf()
```

### Plot - boxenplot

```python
# Create a boxenplot with the Paired palette and the Region column as the hue
sns.boxenplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         palette='Paired',
         hue='Region')

plt.show()
plt.clf()
```

### Plot - barplot

```python
# Show a countplot with the number of models used with each region a different color
sns.countplot(data=df,
         y="Model Selected",
         hue="Region")

plt.show()
plt.clf()
```

### Plot - pointplot

```python
# Create a pointplot and include the capsize in order to show caps on the error bars
sns.pointplot(data=df,
         y='Award_Amount',
         x='Model Selected',
         capsize=.1)

plt.show()
plt.clf()
```

### Plot - countplot

```python
# Create a barplot with each Region shown as a different color
sns.barplot(data=df,
         y='Award_Amount',
         x='Model Selected',
         hue='Region')

plt.show()
plt.clf()
```
