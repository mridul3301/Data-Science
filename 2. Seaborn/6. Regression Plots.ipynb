```python
# import all modules
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read in the DataFrame
df = pd.read_csv("https://assets.datacamp.com/production/repositories/2210/datasets/794e0759b73a2d80baa5d8fb88636a47965139d3/college_datav3.csv")
```

Lets talk about Regression plots and residual plots

*__regplot()__* : shows the relation between features with a fit \
*__residplot__* : Used for evaluating the fit of a model

#### Regression Plot

```python
# Display a regression plot for Tuition
sns.regplot(data=df,
         y='Tuition',
         x='SAT_AVG_ALL',
         marker='^',
         color='g')

plt.show()
plt.clf()
```

```python
# Plot a regression plot of Tuition and the Percentage of Pell Grants
sns.regplot(data=df,
            y='Tuition',
            x='PCTPELL')

plt.show()
plt.clf()
```

#### Residual Plot

```python
# Display the residual plot
sns.residplot(data=df,
          y='Tuition',
          x='SAT_AVG_ALL',
          color='g')

plt.show()
plt.clf()
```

When there are continuous variables, it can be helpful to break them into different bins.

```python
# Create another plot that estimates the tuition by PCTPELL
sns.regplot(data=df,
            y='Tuition',
            x='PCTPELL',
            x_bins=5)

plt.show()
plt.clf()
```

```python
# Create a scatter plot and bin the data into 5 bins
sns.regplot(data=df,
            y='Tuition',
            x='UG',
            x_bins=5)

plt.show()
plt.clf()
```

```python
# Create a regplot and bin the data into 8 bins
sns.regplot(data=df,
         y='Tuition',
         x='UG',
         x_bins=8)

plt.show()
plt.clf()
```

If a value greater than 1 is passed to the *__order__* parameter of regplot(), then Seaborn will attempt a polynomial fit using underlying NumPy functions.

```python
# The final plot should include a line using a 2nd order polynomial
sns.regplot(data=df,
            y='Tuition',
            x='PCTPELL',
            x_bins=5,
            order=2)

plt.show()
plt.clf()
```

```python
# Create a scatter plot by disabling the regression line
sns.regplot(data=df,
            y='Tuition',
            x='UG',
            fit_reg=False)

plt.show()
plt.clf()
```

Seaborn also supports regression plots with categorical variables. 

In some cases, even with the jitter, it might be difficult to see if there are any trends based on the value of the variable. Using an estimator for the x value can provide another helpful view of the data. 

```python
# The final plot should include a line using a 2nd order polynomial using x_estimator
sns.regplot(data=df,
            y='Tuition',
            x='PCTPELL',
            x_bins = 8,
            x_estimator=np.mean,
           order = 2)

plt.show()
plt.clf()
```
