### Importing basic text files

Text files can be broadly classified into two types:\
    1. Plain Text\
    2. CSV Files (Flat Files)

#### Reading a text file:

We use ***open(filename, mode = 'mode')*** to open a text file

```python
filename = 'huck_finn.txt'
file_1 = open(filename, mode = 'r') #r denotes the read mode
```

Assign the text from the file to a variable text by applying the method read 

```python
text_1 = file_1.read()
```

After this, we need to make sure that the connection to the file is closed using the command file dot close.

```python
file_1.close()
```

```python
print(text_1)
```

Using a with statement allows us to create a context in which we can execute commands with the file open. Once out of this clause/context, the file is no longer open and, for this reason, with is called a Context Manager. 

```python
with open('huck_finn.txt', 'r') as my_file:
    print(my_file.read())
```

#### Importing entire text files.

```python
# Open a file: file
file = open('moby_dick.txt', mode = 'r')

# Print it
print(file.read())

```

```python
# Check whether file is closed
print(file.closed)

# Close file
file.close()

# Check whether file is closed
print(file.closed)
```

```python
# Read & print the first 3 lines
with open('moby_dick.txt') as file:
    print(file.readline())
    print(file.readline())
    print(file.readline())
```

### Flat Files

#### Importing Flat Files using Numpy

If all the data are numerical, we can use the package numpy to import the data as a numpy array. 

This is because numpy arrays are the Python standard for storing numerical data. They are efficient, fast and clean. 

We can achieve this using:\
    1. loadtxt()\
    2. genfromtxt()

```python
#start by importing numpy
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#save filename as a variable
filename = 'MNIST.txt'
```

```python
data_MNIST = np.loadtxt(filename, delimiter = ',')
```

```python
data_MNIST
```

```python
# Assign filename to variable: file
file_x = 'MNIST.txt'

# Load file as array: digits
digits = np.loadtxt(file_x, delimiter=',')

# Print datatype of digits
print(type(digits))

# Select and reshape a row
im = digits[21, 1:]
im_sq = np.reshape(im, (28, 28))

# Plot reshaped data (matplotlib.pyplot already loaded as plt)
plt.imshow(im_sq, cmap='Greys', interpolation='nearest')
plt.show()
```

```python
# Assign filename: file
file = 'seaslug.txt'

# Import file: data
data = np.loadtxt(file, delimiter='\t', dtype=str)

# Print the first element of data
print(data[0])

# Import data as floats and skip the first row: data_float
data_float = np.loadtxt(file, delimiter="\t", dtype=float, skiprows=1)

# Print the 10th element of data_float
print(data_float[9])

# Plot a scatterplot of the data
plt.scatter(data_float[:, 0], data_float[:, 1])
plt.xlabel('time (min.)')
plt.ylabel('percentage of larvae')
plt.show()
```
