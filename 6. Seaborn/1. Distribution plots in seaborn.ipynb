*seaborn* is used to create visualizations

*matplotlib* provides the raw building blocks for seaborn's visualization.

```python
# import all modules
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read in the DataFrame
df = pd.read_csv("https://assets.datacamp.com/production/repositories/2210/datasets/205443d734f177d36dad2f0bdf821a57b2c4cc13/schoolimprovement2010grants.csv")
```

```python
df.head()
```

Using Pandas to generate histogram


```python
# Display pandas histogram
df['Award_Amount'].plot.hist()
plt.show()

# Clear out the pandas histogram
plt.clf()
```

Using seaborn to generate Distribution plots.
In this type of plots, we analyze how a single variable is distributed.

Basis for building distribution plots is : *__displot__*

```python
# Display a Seaborn displot
sns.displot(df['Award_Amount'])
plt.show()

# Clear the displot
plt.clf()
```

## Distribution Plot

We can customize the type of plot by using different arguments. In this example, we can use the same displot() function to create a standard histogram and overlay a kernel density element or KDE plot and customize the number of bins to further understand how the values are distributed. This basic approach is used by most functions in Seaborn and is a powerful tool for analyzing data quickly.

```python
# Display a Seaborn displot
sns.displot(df['Award_Amount'], kde = True, bins = 10)
plt.show()

# Clear the displot
plt.clf()
```

setting the rug argument to True adds tickmarks for each observation. 

```python
# Display a Seaborn displot
sns.displot(df['Award_Amount'], kde = True, bins = 10, rug = True, fill = True)
plt.show()

# Clear the displot
plt.clf()
```

### Further Plot Types

The displot() function we have been discussing relies on using additional Seaborn functions such as the kdeplot(),rugplot() and ecdfplot(). By understanding this relationship, we can further customize Seaborn plots by passing additional arguments to the underlying functions.
For Example : Here, we are using an Estimated Cumulative Density Function for our illustration.

```python
# Display a Seaborn displot
sns.displot(df['Award_Amount'], kind = 'ecdf')
plt.show()

# Clear the displot
plt.clf()
```

Let's observe column *Award_Amount*

```python
# Create a displot
sns.displot(df['Award_Amount'],
             bins=20)

# Display the plot
plt.show()
```

```python
# Create a displot of the Award Amount
sns.displot(df['Award_Amount'],
             kind = 'kde',
             rug = True,
             fill = True)

# Plot the results
plt.show()
```
